```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

tesla_data.reset_index(inplace=True)

tesla_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

import yfinance as yf


tesla_financials = tesla.financials


tesla_revenue = tesla_financials.loc['Total Revenue'] 
tesla_revenue_df = pd.DataFrame(tesla_revenue)
tesla_revenue_df.reset_index(inplace=True) 
tesla_revenue_df.columns = ['Date', 'Total Revenue'] 


tesla_revenue_df.tail()
```

```python
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

gme_data.reset_index(inplace=True)

gme_data.head()
```

```python
gme_financials = gme.financials


gme_revenue = gme_financials.loc['Total Revenue'] 
gme_revenue_df = pd.DataFrame(gme_revenue) 
gme_revenue_df.reset_index(inplace=True)  
gme_revenue_df.columns = ['Date', 'Total Revenue']  


gme_revenue_df.tail()
```

```python
import matplotlib.pyplot as plt


def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index.values, data['Close'].values, label='Closing Price') 
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()


make_graph(tesla_data, 'Tesla Stock Price Over Time')
```

```python
make_graph(gme_data, 'GameStop Stock Price Over Time')
```
