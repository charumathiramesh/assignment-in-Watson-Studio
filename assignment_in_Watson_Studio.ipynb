```python
import yfinance as yf

# Fetch Tesla stock data
tsla = yf.Ticker("TSLA")
tsla_data = tsla.history(period="5y")  # Adjust as needed
print(tsla_data.head())  # Display the first few rows
```

Question 2 - Extracting Tesla Revenue Data Using Webscraping - 1 Points

```python
import requests
from bs4 import BeautifulSoup

# URL for Tesla's financials
url = "https://finance.yahoo.com/quote/TSLA/financials?p=TSLA" # Added query parameter to the url

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Check if the page was found
if response.status_code == 200:
    # Extract revenue data (adjust based on the actual HTML structure)
    revenue_element = soup.find('div', attrs={'data-test': 'fin-row'}) # Find the div that contains the revenue data
    if revenue_element:
        revenue = revenue_element.find_all('div', recursive=False)[1].text # Extract the revenue value
        print(f'Tesla Revenue: {revenue}')
    else:
        print("Revenue data not found.")
else:
    print(f"Failed to retrieve page. Status code: {response.status_code}")
```
